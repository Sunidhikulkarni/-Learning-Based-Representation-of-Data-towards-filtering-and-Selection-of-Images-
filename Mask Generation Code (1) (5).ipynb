```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from tkinter import Tk
from tkinter.filedialog import askopenfilename
from tkinter import filedialog
from tkinter import *
import os
from os import listdir
from os.path import isfile, join
```

```python
def shape_selection(event, x, y, flags, param):
    global ref_point, crop
    if event == cv2.EVENT_LBUTTONDOWN:
        ref_point = [(x, y)]
    elif event == cv2.EVENT_LBUTTONUP:
        ref_point.append((x, y))
        cv2.rectangle(image, ref_point[0], ref_point[1], (0, 255, 0), 2)
        cv2.imshow("image", image)
```

```python
flag = 0
root = Tk()
root.withdraw()
folder_selected = filedialog.askdirectory()
path = folder_selected + "_mask"
isdir = os.path.isdir(path)  
if isdir == False:
    os.mkdir(path)
onlyfiles = [f for f in listdir(folder_selected) if isfile(join(folder_selected, f))]

i = 0
while i<len(onlyfiles):
    if flag == 1:
        flag = 0
        i = i - 1
    print("Image ",i+1,"out of ",str(len(onlyfiles)),end = " ,")
    print((((i+1)/len(onlyfiles))*100), "% Completed")
    filename = folder_selected + "/" + onlyfiles[i]
    image = cv2.imread(filename)
    image = cv2.resize(image, (299,299),interpolation = cv2.INTER_AREA)
    clone = image.copy()
    cv2.namedWindow("image")
    cv2.setMouseCallback("image", shape_selection)
    while True:
        cv2.imshow("image", image)
        key = cv2.waitKey(1) & 0xFF
        if key == ord("r"):
            image = clone.copy()
        elif key == ord("q"):
            break
        elif key == ord("z"):
            flag = 2
            break
    cv2.destroyAllWindows() 
    mask = np.zeros(image.shape[:2],np.uint8)
    bgdModel = np.zeros((1,65),np.float64)
    fgdModel = np.zeros((1,65),np.float64)
    rect = (ref_point[0][0],ref_point[0][1],ref_point[1][0],ref_point[1][1])
    cv2.grabCut(clone,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)

    mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
    clone = clone*mask2[:,:,np.newaxis]

    cv2.imshow('image',clone)
    k = cv2.waitKey(0)
    if k == ord('q'):
        cv2.destroyAllWindows()
        name = path + "/" + onlyfiles[i][:-4] + "_mask.jpg"
        cv2.imwrite(name,clone)
    elif k == ord('r'):
        flag = 1
    elif k == ord('z'):
        i = len(onlyfiles) - 1
        cv2.destroyAllWindows()
    i+=1
```
